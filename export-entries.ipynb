```python
with open('/Users/baart_f/src/bibliography/sealevel.bib', 'w') as f:
    for entry in entries:
        bib_entry = bibs.entries[entry]
        remove_keys = [key for key in bib_entry.fields if '.' in key]
        for key in remove_keys:
            del bib_entry.fields[key]
        f.write(bib_entry.to_string('bibtex'))
```
